```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("../data/Chevron Challenge Materials/training.xlsx",index_col=0)
```

```python
df.head()
```

```python
# Set a clean and minimalistic style
sns.set(style="ticks")

plt.figure(figsize=(10, 6))  # Larger figure for better readability
sns.countplot(x=df['Vehicle Category'], palette='coolwarm_r',edgecolor="black", linewidth=2)  # Keep the palette for visual appeal

# Change the background color to light beige
plt.gcf().set_facecolor('#f4e1d2')  # Light brown color
plt.gca().set_facecolor('#f4e1d2')  # Light brown color for the plot area

# Refined plot title and labels
plt.title('Vehicle Category Distribution', fontsize=16, color='darkblue', weight='normal')
plt.xlabel("Vehicle Categories", fontsize=14, color='black', weight='normal')
plt.ylabel("Counts", fontsize=14, color='black', weight='normal')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45, ha='right', fontsize=12)


plt.grid(True, linestyle='--', alpha=0.5)


for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', 
                       fontsize=12, color='black', 
                       xytext=(0, 8), textcoords='offset points')

# Tight layout for a clean finish
plt.tight_layout()
plt.show()
```

From our findings, the category for T6 carries more weightage over the time frame and is probably a wide consumer choice

```python
# Count fuel type occurrences per year
fuel_trend = df.groupby(["Date", "Fuel Type"]).size().unstack().fillna(0)
# Set a clean, minimalistic style
sns.set(style="ticks")

# Plot the trend of fuel type distribution over time
plt.figure(figsize=(15, 6))

# Use a color palette and plot with markers
sns.lineplot(data=fuel_trend, markers='o', palette='coolwarm_r', linewidth=5)

# Change the background color to light beige
plt.gcf().set_facecolor('#f4e1d2')  # Light brown color
plt.gca().set_facecolor('#f4e1d2')  # Light brown color for the plot area

# Refined title and labels with minimal styling
plt.title("Trend of Fuel Type Distribution Over Time", fontsize=16, color='darkblue', weight='normal')
plt.xlabel("Year", fontsize=14, color='black', weight='normal')
plt.ylabel("Count of Vehicles", fontsize=14, color='black', weight='normal')

# Customize legend appearance
plt.legend(title="Fuel Type", fontsize=12, title_fontsize=14, loc='upper left', frameon=False)

# Optional: Add a subtle grid for context
plt.grid(True, linestyle='--', alpha=0.5)

# Tight layout for a clean finish
plt.tight_layout()
plt.show()
```
