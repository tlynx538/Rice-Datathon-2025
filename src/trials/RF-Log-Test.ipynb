```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error
from scipy.stats import boxcox
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("../data/processed/removed-unn-corr.csv",index_col=0)
```

```python
X = df.drop(['Vehicle Population'],axis=1)
```

```python
y = df['Vehicle Population']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1675)
```

```python
y_train, lambda_val = boxcox(y_train + 1) 
```

```python
model = RandomForestRegressor(max_depth=50, n_estimators=150, random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = (y_pred * lambda_val) + 1
```

```python
root_mean_squared_error(y_test, y_pred)
```
