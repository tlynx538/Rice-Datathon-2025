```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/cleaned.csv")
```

```python
data.head()
```

```python
# Convert to datetime and extract year
#data['Date'] = pd.to_datetime(data['Date']).dt.year

# Group by Year
yearly_data = data.groupby('Date')['Vehicle Population'].sum().reset_index()

# Plot yearly trends
plt.figure(figsize=(10, 6))
plt.plot(yearly_data['Date'], yearly_data['Vehicle Population'], 
         marker='o', linestyle='--', color='#2ecc71')
plt.title('Vehicle Population Trends by Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Vehicle Population', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(yearly_data['Date'])
plt.tight_layout()
plt.show()
```
